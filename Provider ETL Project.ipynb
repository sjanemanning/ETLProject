```python
import pandas as pd
from sqlalchemy import create_engine
csv1= "FY2015.csv"
FY2015 = pd.read_csv(csv1)
```

```python
FY2015.columns
```

```python
FY2015=FY2015[['DRG Definition', 'Provider Id', 'Provider Name','Provider State','Hospital Referral Region (HRR) Description',
       'Total Discharges', 'Average Covered Charges', 'Average Total Payments',
       'Average Medicare Payments']]
FY2015.head()
```

```python
FY2015=FY2015.rename(columns={'DRG Definition':'drg_definition', 'Provider Id':'provider_id', 'Provider Name':'provider_name','Provider State':'provider_state','Hospital Referral Region (HRR) Description':'hrr','Total Discharges':'discharges', 'Average Covered Charges':'covered_charges', 'Average Total Payments':'payments','Average Medicare Payments':'medicare_payments'})       
FY2015['year']=2015                  
FY2015.head()
```

```python
csv2= "FY2016.csv"
FY2016 = pd.read_csv(csv2)
FY2016=FY2016[['DRG Definition', 'Provider Id', 'Provider Name','Provider State','Hospital Referral Region (HRR) Description',
       'Total Discharges', 'Average Covered Charges', 'Average Total Payments',
       'Average Medicare Payments']]
FY2016=FY2016.rename(columns={'DRG Definition':'drg_definition', 'Provider Id':'provider_id', 'Provider Name':'provider_name','Provider State':'provider_state','Hospital Referral Region (HRR) Description':'hrr','Total Discharges':'discharges', 'Average Covered Charges':'covered_charges', 'Average Total Payments':'payments','Average Medicare Payments':'medicare_payments'})
FY2016['year']=2016    
```

```python
engine = create_engine('postgres://postgres:oklahoma@localhost:5432/ProviderUtilization')
connection = engine.connect()
```

```python
FY2015.to_sql(name='inpatientutilization', con=engine, if_exists='append', index=False)
```

```python
FY2016.to_sql(name='inpatientutilization', con=engine, if_exists='append', index=False)
```
